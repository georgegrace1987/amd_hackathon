```python
import duckdb

DB_PATH = "complaints.duckdb"

def get_conn():
    return duckdb.connect(DB_PATH)

def setup_user_authentication():
    conn = get_conn()
    
    # 1. Drop the table if it already exists to ensure a clean slate
    conn.execute("DROP TABLE IF EXISTS user_authentication;")
    print("Dropped existing 'user_authentication' table (if any).")
    
    # 2. Re-create the user_authentication table
    conn.execute("""
        CREATE TABLE user_authentication (
            UserID    VARCHAR PRIMARY KEY,
            Password  VARCHAR NOT NULL,
            UserType  VARCHAR NOT NULL,
            Team      VARCHAR NOT NULL
        )
    """)
    print("Created 'user_authentication' table.")
    
    # 3. Define the specific users (4 values per tuple)
    specific_users = [
        ("FraudMember01", "FM01", "TeamMember", "Fraud"),
        ("FraudMember02", "FM02", "TeamMember", "Fraud"),
        ("FraudSupervisor01", "FMS01", "Supervisor", "Fraud"),
        ("KYCMember01", "KYC01", "TeamMember", "KYC"),
        ("KYCSupervisor01", "KYCS01", "Supervisor", "KYC"),
        ("InsuranceMember01", "IM01", "TeamMember", "Insurance"),
        ("InsuranceSupervisor01", "IMS01", "Supervisor", "Insurance"),
        ("CorporateMember01", "CM01", "TeamMember", "Corporate"),
        ("CorporateSupervisor01", "CMS01", "Supervisor", "Corporate"),
        ("Admin", "Admin", "Admin", "Admin")
    ]
    
    # 4. Insert the users (Fixed to include the 'Team' column)
    conn.executemany("""
        INSERT INTO user_authentication (UserID, Password, UserType, Team)
        VALUES (?, ?, ?, ?)
    """, specific_users)
    print(f"Inserted {len(specific_users)} users into 'user_authentication'.")
    
    # 5. Verify the data was inserted correctly
    result = conn.execute("SELECT * FROM user_authentication").fetchall()
    print("\n--- Current Table Data ---")
    for row in result:
        print(row)
        
    conn.close()

if __name__ == "__main__":
    setup_user_authentication()
```
